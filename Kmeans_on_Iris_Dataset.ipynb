```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/IRIS.csv')
df
```

```python
from sklearn.cluster import KMeans
km=KMeans()
sse=[]
k_rng=range(1,10)
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['sepal_length','sepal_width','petal_length','petal_width']])
  sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum os squared error')
plt.plot(k_rng,sse)
```

```python
x = df.iloc[:, [0, 1, 2, 3]].values
```

```python
km=KMeans(n_clusters=3)
km.fit(x)
```

```python
y_p=km.fit_predict(df[["sepal_length","sepal_width","petal_length","petal_width"]])
y_p
```

```python
km.cluster_centers_
x=km.cluster_centers_[:,0]
y=km.cluster_centers_[:,1]
z=km.cluster_centers_[:,2]
print(x)
print(y)
```
